```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import SMOTE

results_knn=[]
for gene in DEG:
    if gene not in x_scl.columns:
        continue
    x_gene=x_scl[[gene]]
    acc_arr=[]

    for tr_id, te_id in kf.split(x_gene):
        x_tr, x_te = x_gene.iloc[tr_id], x_gene.iloc[te_id]
        y_tr, y_te = y.iloc[tr_id], y.iloc[te_id]
        smote = SMOTE(random_state=42)
        x_tr_resample, y_tr_resample = smote.fit_resample(x_tr, y_tr)

        knn = KNeighborsClassifier(n_neighbors=5)
        knn.fit(x_tr_resample, y_tr_resample)
        #y_prob = knn.predict_proba(x_te)[:,1]
        y_pred = knn.predict(x_te)

        acc = accuracy_score(y_te, y_pred)
        acc_arr.append(acc)

    results_knn.append((gene,(sum(acc_arr)/len(acc_arr))))
results_knn.sort(key=lambda x: x[1], reverse=True)

for gene, acc_arr in results_knn[:5]:
    print(f"{gene}: acc_arr={acc_arr:.3f}")
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
results_rf=[]
for gene in DEG:
    if gene not in x_scl.columns:
        continue
    x_gene=x_scl[[gene]]
    acc_arr=[]

    for tr_id, te_id in kf.split(x_gene):
        x_tr, x_te = x_gene.iloc[tr_id], x_gene.iloc[te_id]
        y_tr, y_te = y.iloc[tr_id], y.iloc[te_id]

        smote = SMOTE(random_state=42)
        x_tr_resample, y_tr_resample = smote.fit_resample(x_tr,y_tr)

        rf = RandomForestClassifier(n_estimators=5, max_depth=None, random_state=42)
        rf.fit(x_tr_resample, y_tr_resample)
        y_prob = rf.predict_proba(x_te)[:,1]
        y_pred = rf.predict(x_te)

        acc = accuracy_score(y_te, y_pred)
        acc_arr.append(acc)

    results_rf.append((gene,(sum(acc_arr)/len(acc_arr))))
results_rf.sort(key=lambda x: x[1], reverse=True)

for gene, acc_arr in results_rf[:5]:
    print(f"{gene}: acc_arr={acc_arr:.3f}")
```

```python
import matplotlib.pyplot as plt

scores = [score for _, score in results_rf]
chunk_size = 15
group_avgs = [np.mean(scores[i:i + chunk_size]) for i in range(0, len(scores), chunk_size)]

x_values = list(range(chunk_size, len(scores) + chunk_size, chunk_size))
y_values = [avg * 100 for avg in group_avgs]

x_at_80 = None
for i in range(len(y_values)):
    if y_values[i] < 80:
        x_at_80 = x_values[i]
        break

plt.figure(figsize=(10, 6))
plt.step(x_values, y_values, color='red', where='post', label='Average ROC AUC (%)')
plt.axhline(y=80, color='blue', linestyle='--', label='80% Threshold')
plt.axvline(x=x_at_80, color='blue', linestyle='--', label='x at 80%')
plt.xlabel('Gene Index in Sorted Gene Set', fontsize=12)
plt.ylabel('Average ROC AUC (%)', fontsize=12)
plt.title('(a) GSE13601', fontsize=14)
plt.ylim(40, 105)
plt.legend()
plt.grid(True)
plt.show()
```
