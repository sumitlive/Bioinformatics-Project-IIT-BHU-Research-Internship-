```python
import pandas as pd
import numpy as np
deg_df = pd.read_csv('(UPDATED)DEG_expression_matrix_GSE34105.csv')
deg_df
```

```python
from sklearn.model_selection import train_test_split
x = deg_df.iloc[:,0:-1]
x
```

```python
y=deg_df.iloc[:, -1]
y
```

```python
gene = x.columns
max_value = x[gene].max().max()
if max_value < 20:
    print('already transformed')
else:
    x[gene] = np.log2(x[gene])
x
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
x_scl = pd.DataFrame(std.fit_transform(x), index=x.index, columns=x.columns)
x_scl
```

```python
log_fc = pd.read_csv('(UPDATED)DEG_logFC_GSE34105.csv')
DEG = log_fc['ID'].tolist()
print(len(DEG))
```

```python
df=x_scl[[DEG[0]]]
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

feature = df.iloc[:, 0]
labels = y

plt.figure(figsize=(8, 4))
sns.kdeplot(feature[labels == 0], fill=True, color='blue', label='Class 0')
sns.kdeplot(feature[labels == 1], fill=True, color='red', label='Class 1')
plt.xlabel(f"Feature: {df.columns[0]}")
plt.title("Density Plot of Feature by Class")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

# How SMOTE works

```python
from imblearn.over_sampling import SMOTE

gene = DEG[0]
x_gene = x_scl[[gene]]
i=0
for tr_id, te_id in kf.split(x_gene):
  i+=1
  x_tr,x_te = x_gene.iloc[tr_id], x_gene.iloc[te_id]
  y_tr, y_te = y.iloc[tr_id], y.iloc[te_id]
  num_class_0_tr = (y_tr==0).sum()
  num_class_0_te = (y_te==0).sum()
  num_class_1_tr = (y_tr==1).sum()
  num_class_1_te = (y_te==1).sum()
  smote=SMOTE(random_state=42)
  x_tr_resample, y_tr_resample = smote.fit_resample(x_tr, y_tr)
  num_class_0_tr_resample = (y_tr_resample==0).sum()
  num_class_1_tr_resample = (y_tr_resample==1).sum()
  print(f"{i}: num_class_0_tr: {num_class_0_tr}, num_class_0_tr_resample: {num_class_0_tr_resample}, num_class_0_te: {num_class_0_te}")
  print(f"{i}: num_class_1_tr: {num_class_1_tr}, num_class_1_tr_resample: {num_class_1_tr_resample}, num_class_1_te: {num_class_1_te}")
  print()
```
