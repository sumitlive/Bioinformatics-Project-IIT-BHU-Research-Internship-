# Avg MI score

```python
import pandas as pd 
import numpy as np
x1 = pd.read_csv('feature_with_mi_scores_svm.csv',index_col=0)
svm_mi = {}
for i in range(x1.shape[0]):
    a = x1.iloc[i,0]
    b = x1.iloc[i,1]
    svm_mi[a]=b
svm_mi = sorted(svm_mi.items(), key=lambda x:x[0])
svm_mi
```

```python
import pandas as pd 
import numpy as np
x2_ = pd.read_csv('feature_with_mi_scores_randomforest.csv',index_col=0)
rf_mi = {}
for i in range(x2_.shape[0]):
    a = x2_.iloc[i,0]
    b = x2_.iloc[i,1]
    rf_mi[a]=b
rf_mi = sorted(rf_mi.items(), key=lambda x:x[0])
rf_mi
```

```python
import pandas as pd 
import numpy as np
x2 = pd.read_csv('feature_with_mi_scores_xgboost.csv',index_col=0)
xgboost_mi = {}
for i in range(x2.shape[0]):
    a = x2.iloc[i,0]
    b = x2.iloc[i,1]
    xgboost_mi[a]=b
xgboost_mi = sorted(xgboost_mi.items(), key=lambda x:x[0])
xgboost_mi
```

```python
avg_mi = {}
for i in range(130):
    avg_mi[i] = (svm_mi[i][1]+xgboost_mi[i][1]+rf_mi[i][1])/3
avg_mi
```

```python
# values = list(avg_mi.values())
# min_val = min(values)
# max_val = max(values)

# if max_val == min_val:
#     scaled_values = [0 for _ in values] 
# else:
#     scaled_values = [(x - min_val) / (max_val - min_val) for x in values]

# scaled_avg_mi = dict(zip(avg_mi.keys(), scaled_values))
# scaled_avg_mi



values = list(avg_mi.values())

mean_val = sum(values) / len(values)
std_val = (sum((x - mean_val) ** 2 for x in values) / len(values)) ** 0.5

if std_val == 0:
    scaled_values = [0 for _ in values]  # If std is zero, scale to 0
else:
    scaled_values = [(x - mean_val) / std_val for x in values]

scaled_avg_mi = dict(zip(avg_mi.keys(), scaled_values))

scaled_avg_mi
```

# Plot

```python
import matplotlib.pyplot as plt


keys = list(scaled_avg_mi.keys())
values1 = list(scaled_avg_mi.values())
values2 = list(scaled_avg_rfe.values())

plt.figure(figsize=(16, 8))

plt.plot(keys, values1, label='MI Scores', color='#1f77b4', marker='o')
plt.plot(keys, values2, label='RFE Scores', color='#ff7f0e', marker='o')
plt.xlabel('Features')
plt.ylabel('Scores')
plt.title('Comparison of Scores')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig("Line Graph Plot for mi and rfe score.png", dpi=300, bbox_inches="tight")

plt.show()
```
