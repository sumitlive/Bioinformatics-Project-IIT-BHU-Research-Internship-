```python
import pandas as pd
import numpy as np
```

```python
# df_features = pd.read_csv('Features_130.csv',index_col=0)
df_features = pd.read_csv('Features_130_EEMD.csv',index_col=0)
df_features
```

```python
missing_values = df_features.isnull().sum().tolist()
```

```python
def missing_value_function(x):
    index = []
    for i in range(len(x)):
        if missing_values[i] !=0:
            index.append(i)
    return index
missing_values_index = missing_value_function(missing_values)
print(missing_values_index)
```

```python
# column_27 = df_features.iloc[:,27]
# column_27_mean = column_27.mean()
# df_features.iloc[:,27].fillna(column_27_mean, inplace=True)

for i in range(len(missing_values_index)):
  column_mean = df_features.iloc[:,missing_values_index[i]].mean()
  df_features.iloc[:,missing_values_index[i]].fillna(column_mean,inplace=True)
```

```python
df_features.isnull().sum().value_counts()
```

```python
label_df = pd.read_csv('Labels_df.csv', index_col=0)
label_df
```

```python
groups = np.load('groups.npy')
unique_groups = np.unique(groups)
print(unique_groups, len(unique_groups))
```

```python
x = np.array(df_features)
y = np.array(label_df['Valence_Class'])

group_labels = np.array([y[np.where(groups==grp)[0][0]] for grp in unique_groups])
print(group_labels, len(group_labels))
```

```python
print(x.shape)
print(y.shape)
print(len(groups))
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

print(Counter(y))
smote= SMOTE(sampling_strategy='auto',random_state=42)
x_resampled , y_resampled = smote.fit_resample(x,y)

print(Counter(y_resampled))
print(x.shape[0])
print(x_resampled.shape[0])
```

```python
num_synthetic = len(x_resampled)-len(x)
max_group = max(groups)
synthetic_groups = np.arange(max_group + 1, max_group + 1 + num_synthetic)
groups_resampled = np.concatenate([groups,synthetic_groups])
print(len(groups_resampled))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import StratifiedKFold
from tqdm import tqdm
sgkf = StratifiedKFold(n_splits=5, shuffle=True, random_state=206)

fold_accuracies = []
fold_f1_scores = []
all_test = []
all_pred = []
x = x_resampled
y = y_resampled
groups = groups_resampled
for fold_idx, (train_idx, test_idx) in tqdm(enumerate(sgkf.split(x,y,groups))):
    x_train, y_train = x[train_idx], y[train_idx]
    x_test, y_test = x[test_idx], y[test_idx]

    scaler = StandardScaler()
    x_train_scl = scaler.fit_transform(x_train)
    #x_test_scl = scaler.transform(x_test)

    feature_selector = SelectFromModel(RandomForestClassifier(verbose=0, random_state=26), threshold="median" )
    feature_selector.fit(x_train_scl,y_train)
    selected_feature_indices = feature_selector.get_support(indices=True)
    print("Indices of selected features:", selected_feature_indices)

    x_train_selected = x_train_scl[:, selected_feature_indices]
    x_test_selected = x_test[:, selected_feature_indices]

    classifier = RandomForestClassifier(random_state=42)
    classifier.fit(x_train_selected, y_train)

    y_pred = classifier.predict(x_test_selected)
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='macro')
    print(f"Accuracy: {acc * 100:.2f}%")
    print(f"F1 Score: {f1 * 100:.4f}%")

    fold_accuracies.append(acc)
    fold_f1_scores.append(f1)
    all_test.extend(y_test)
    all_pred.extend(y_pred)

print("\n=== Overall Results ===")
print(f"Mean Accuracy: {np.mean(fold_accuracies) * 100:.2f}%")
print(f"Standard Deviation: {np.std(fold_accuracies):.4f}")
print(f"Mean F1-score: {np.mean(fold_f1_scores) * 100:.4f}%")
print("Classification Report:\n", classification_report(all_test, all_pred))
```
