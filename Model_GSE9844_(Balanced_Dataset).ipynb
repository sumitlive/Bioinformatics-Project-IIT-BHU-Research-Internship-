```python
import pandas as pd
import numpy as np
deg_df = pd.read_csv('(UPDATED)DEG_expression_matrix_GSE9844.csv')
deg_df
```

```python
from sklearn.model_selection import train_test_split
x = deg_df.iloc[:,0:-1]
x
```

```python
y=deg_df.iloc[:, -1]
y
```

```python
gene = x.columns
max_value = x[gene].max().max()
if max_value < 20:
    print('already transformed')
else:
    x[gene] = np.log2(x[gene])
x
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
x_scl = pd.DataFrame(std.fit_transform(x), index=x.index, columns=x.columns)
x_scl
```

```python
log_fc = pd.read_csv('(UPDATED)DEG_logFC_GSE9844.csv')
DEG = log_fc['ID'].tolist()
print(len(DEG))
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import SMOTE

results_knn=[]
for gene in DEG:
    if gene not in x_scl.columns:
        continue
    x_gene=x_scl[[gene]]
    acc_arr=[]

    for tr_id, te_id in kf.split(x_gene):
        x_tr, x_te = x_gene.iloc[tr_id], x_gene.iloc[te_id]
        y_tr, y_te = y.iloc[tr_id], y.iloc[te_id]
        smote = SMOTE(random_state=42)
        x_tr_resample, y_tr_resample = smote.fit_resample(x_tr, y_tr)

        knn = KNeighborsClassifier(n_neighbors=5)
        knn.fit(x_tr_resample, y_tr_resample)
        y_prob = knn.predict_proba(x_te)[:,1]
        y_pred = knn.predict(x_te)

        acc = accuracy_score(y_te, y_pred)
        acc_arr.append(acc)

    results_knn.append((gene,(sum(acc_arr)/len(acc_arr))))
results_knn.sort(key=lambda x: x[1], reverse=True)

for gene, acc_arr in results_knn[:5]:
    print(f"{gene}: acc_arr={acc_arr:.3f}")
```

```python
df = pd.DataFrame(results_knn, columns=['gene_id','knn_acc_score'])
df.to_csv('knn_acc_score_balanced_dataset_GSE9844.csv', index=False)
```

# SVM

```python
from sklearn.svm import SVC
results_svm=[]
acc_svm=[]

for gene in DEG:
    if gene not in x_scl.columns:
        continue
    x_gene=x_scl[[gene]]
    acc_arr=[]

    for tr_id, te_id in kf.split(x_gene):
        x_tr, x_te = x_gene.iloc[tr_id], x_gene.iloc[te_id]
        y_tr, y_te = y.iloc[tr_id], y.iloc[te_id]

        smote = SMOTE(random_state=42)
        x_tr_resample, y_tr_resample = smote.fit_resample(x_tr,y_tr)

        svm = SVC(kernel='rbf', probability=True, random_state=42)
        svm.fit(x_tr_resample, y_tr_resample)
        y_prob = svm.predict_proba(x_te)[:,1]
        y_pred = svm.predict(x_te)

        acc = accuracy_score(y_te, y_pred)
        acc_arr.append(acc)
    results_svm.append((gene,(sum(acc_arr)/len(acc_arr))))
results_svm.sort(key=lambda x: x[1], reverse=True)

for gene, acc_arr in results_svm[:5]:
    print(f"{gene}: acc_arr={acc_arr:.3f}")
```

```python
df = pd.DataFrame(results_svm, columns=['gene_id','svm_acc_score'])
df.to_csv('svm_acc_score_balanced_dataset_GSE9844.csv', index=False)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
results_rf=[]
for gene in DEG:
    if gene not in x_scl.columns:
        continue
    x_gene=x_scl[[gene]]
    acc_arr=[]

    for tr_id, te_id in kf.split(x_gene):
        x_tr, x_te = x_gene.iloc[tr_id], x_gene.iloc[te_id]
        y_tr, y_te = y.iloc[tr_id], y.iloc[te_id]

        smote = SMOTE(random_state=42)
        x_tr_resample, y_tr_resample = smote.fit_resample(x_tr,y_tr)

        rf = RandomForestClassifier(n_estimators=5, max_depth=None, random_state=42)
        rf.fit(x_tr_resample, y_tr_resample)
        y_prob = rf.predict_proba(x_te)[:,1]
        y_pred = rf.predict(x_te)

        acc = accuracy_score(y_te, y_pred)
        acc_arr.append(acc)

    results_rf.append((gene,(sum(acc_arr)/len(acc_arr))))
results_rf.sort(key=lambda x: x[1], reverse=True)

for gene, acc_arr in results_rf[:5]:
    print(f"{gene}: acc_arr={acc_arr:.3f}")
```

```python
df = pd.DataFrame(results_rf, columns=['gene_id','rf_acc_score'])
df.to_csv('rf_acc_score_balanced_dataset_GSE9844.csv', index=False)
```

# Logistic Regression
